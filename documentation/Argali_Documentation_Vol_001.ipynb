# Argali Documentation
Argali is a data analytics framework that provides high level access to a wide range of statistics and visualisations. This package can be used to develop an understanding of variables and data sets. Argali can also be used to create summaries of variables that can provide consistent overviews for analytical process'
## Univariate Analysis
The Argali univariate analysis provides a descriptive overview of a single variable with a wide range of easily accessible statistics and graphs. 

Argali's univariate summary report provides statistical figures and their interpretations along with detailed grpahical representations. This summary provides a strong base for exploratory analysis and enables the user to quickly access the desired visualisation while simultationsly building confidence in the interpretation through detailed commentary specific to the variable.

Additionally the Argali univariate analysis allow access to all individual statistics based on the most commonly used name. A list of statistics can be found in the inidividual statistics section (19 statistics at present). There are four plot grids that are also accessible on an individual bases to enhance the ability of the use to quickly encorporate visualisations into report.

The ability to access individual stats means that the user is able to examine the summary report and identify key statistics and visualisations relevent to their requirements. Enabling the creation of bespoke reports in a quick and efficient manner.

#### Import

```python
from argali import descriptive_statistics as DS
```

#### Load data

```python
x = [1,9,2,2,8,8,3,3,3,7,7,7,4,4,4,4,6,6,6,6,5,5,5,5,5,4,2,7,6,5,9,1,2,3,4,5,6,7,8,9,1,2,3,4,5,6,7,8,9,1,2,3,4,3,4,3,4,5,4,5,4,5,5,5,5,5,6,7,6,7,6,8,8,8,8,8,8,7,7,7,7,6,6,5,5,4,4]
```

#### Create univariate analysis object

```python
x_analysis = DS.Univariate(x)
```

#### Create Summary Report

The summary report includes a variety of measures of central tendency and dispersion. Additionally skew and kurtosis are calculated. An example report can be seen below:

```python
x_analysis.descriptive_summary()
```

All of the components in the above report can be accessed individually and assigned to new variables. The directory command provides a list of all the possible statistics and plots that can be accessed. The following section demonstrates how to access each individual statistic and plot.

```python
dir(x_analysis)
```
